```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Generate sample data
from sklearn.datasets import make_blobs

# Generate synthetic dataset with 3 centers and some noise
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
# Standardize the dataset (important for DBSCAN performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply DBSCAN clustering
db = DBSCAN(eps=0.3, min_samples=10)
db.fit(X_scaled)

# Get the labels (cluster assignments)
labels = db.labels_
```

```python
# Extract noise points (label -1)
noise_points = X[labels == -1]

# Extract clustered points
clustered_points = X[labels != -1]
```

```python
# Plotting the results
plt.figure(figsize=(8, 6))

# Plot clustered points
plt.scatter(clustered_points[:, 0], clustered_points[:, 1], c=labels[labels != -1], cmap='viridis', marker='o', label='Clusters')

# Plot noise points
plt.scatter(noise_points[:, 0], noise_points[:, 1], color='red', marker='x', label='Noise')

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Get unique cluster labels (excluding noise)
unique_labels = set(labels)
unique_labels.discard(-1)  # Remove noise label

# Estimate cluster sizes and print
for label in unique_labels:
    cluster_size = np.sum(labels == label)
    print(f"Cluster {label}: {cluster_size} points")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN clustering
db = DBSCAN(eps=0.3, min_samples=10)
db.fit(X_scaled)

# Get labels and classify noise points
labels = db.labels_

# Extract noise points (label -1)
noise_points = X[labels == -1]

# Extract clustered points
clustered_points = X[labels != -1]

# Plotting the results
plt.figure(figsize=(8, 6))

# Plot clustered points
plt.scatter(clustered_points[:, 0], clustered_points[:, 1], c=labels[labels != -1], cmap='viridis', marker='o', label='Clusters')

# Plot noise points
plt.scatter(noise_points[:, 0], noise_points[:, 1], color='red', marker='x', label='Noise')

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Estimate cluster sizes
unique_labels = set(labels)
unique_labels.discard(-1)  # Remove noise label

for label in unique_labels:
    cluster_size = np.sum(labels == label)
    print(f"Cluster {label}: {cluster_size} points")
```
